#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
filtr1 = ds["level5"] == "Valdemorillo"
ds_def1 = ds[filtr1]
promd1 = ds_def1["price"].mean()
print("El promedio de los precios pertecientes a los inmuebles de la población de Valdemorillo es de:", math.ceil(promd1), "€.")
print()

filtr2 = ds["level5"] == "Galapagar"
ds_def2 = ds[filtr2]
promd2 = ds_def2["price"].mean()
print("El promedio de los precios pertecientes a los inmuebles de la población de Galapagar es de:", math.ceil(promd2), "€.")
print()

if promd1 == promd2:
    print("Efectivamente, son iguales.")
else:
    print("No son iguales.")
```

#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
ds_def1 = ds_def1.copy()
ds_def1["pps"] = ds_def1["price"] / ds_def1["surface"]
prm1 = ds_def1["pps"].mean()
print("El promedio de los precios por metro cuadrado pertecientes a los inmuebles de la población de Valdemorillo es de:", math.ceil(prm1), "€.")
print()

ds_def2 = ds_def2.copy()
ds_def2["pps"] = ds_def2["price"] / ds_def2["surface"]
prm2 = ds_def2["pps"].mean()
print("El promedio de los precios por metro cuadrado pertecientes a los inmuebles de la población de Galapagar es de:", math.ceil(prm2), "€.")
print()

if prm1 == prm2:
    print("Efectivamente, son iguales.")
else:
    print("No son iguales.")
```

#### Ejercicio 11. Analiza la relación entre la superficie y el precio de las casas. (★★☆)

Pista: Puedes hacer un `scatter plot` y luego escribir una conclusión al respecto.

```python
plt.figure(figsize=(12, 6))
plt.scatter(ds['price'], ds['surface'], color='blue', alpha=0.6, edgecolors='black')

plt.xscale('log')
plt.yscale('log')
plt.xlabel("Superficie (m²)", fontsize=12)
plt.ylabel("Precio del Inmueble (€)", fontsize=12)
plt.title("Relación entre Superficie y Precio de Bienes Inmuebles", fontsize=15)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

Los valores se representaban ineficazmente si empleaba una escala normal, la escala logarítmica logra alcanzar una mayor claridad, existe una relación entre ambas columnas, de forma que mientras más vaya aumentando la superficie del bien inmueble, más se incrementará el precio de la vivienda.
